```python
import statsmodels.formula.api as smf
model=smf.ols("sunday~daily",data=newspaper).fit()
```

```python
sns.regplot(x="daily",y="sunday",data=newspaper);
```

```python
model.params
```

```python
model.fittedvalues
```

```python
newspaper['predicted1']=13.835630+ 1.339715*newspaper['daily']
newspaper
```

```python
newspaper['error1']=newspaper['sunday']-newspaper['predicted1']
newspaper
```

```python
newspaper['predicted']=model.fittedvalues
newspaper
```

```python
newspaper['error']=model.resid
newspaper
```

```python
newspaper['error-sq']=(newspaper['error'])**2
newspaper
```

```python
mean_squared_error=newspaper['error-sq'].mean()
mean_squared_error
```
